```python
a = os.listdir('/drive/MyDrive/Doutorado_v2/Files_in_PDF')
df_files = pd.DataFrame(list(a), columns=['name_file'])
df_files.head()
```

```python
df_files.shape
```

```python
df_files['data_noticia_orig'] = 'a'
df_files['Publication_orig'] = 'a'
df_files['titulo'] = 'a'
df_files['Size_notice'] = 0
df_files['Sumarization'] = 'a'
df_files.head()
```

```python
def chr_remove(old, to_remove):
    new_string = old
    for x in to_remove:
        new_string = new_string.replace(x, '')
    return new_string
```

```python
path_in = '/drive/MyDrive/Doutorado_v2/Files_in_PDF'
```

```python
row_index = 0
for k in range(0, len(df_files)):
    name_file_df = df_files.iloc[k]['name_file']
    file_in = os.path.join(path_in, name_file_df)

    if os.stat(file_in).st_size == 0:
        row_index+=1
    else:
        text = ''
        with fitz.open(file_in) as doc:
            for page in doc:
                text+= page.get_text()

        page001 = text.split("\n")[1:2]
        page002 = text.split("\n")[2:3]
        page003 = text.split("\n")[3:4]

        page004 = text.split("\n")[4:]
        page004 = repr(page004)
        page004 = chr_remove(page004, "',")
        page004 = chr_remove(page004, '"')
        page004 = " ".join(page004.split())

        df_files.loc[row_index, 'data_noticia_orig'] = page001[0]
        df_files.loc[row_index, 'Publication_orig'] = page002[0]
        df_files.loc[row_index, 'titulo'] = page003[0]
        df_files.loc[row_index, 'Size_notice'] = len(page004)

        row_index+=1
```

# Copy from Dataframe

```python
df_files2 = df_files.copy()
```

```python
df_files2[(df_files2['Size_notice'] == 0)]
```

```python
df_files2 = df_files2[(df_files2['Size_notice'] != 0)]
df_files2.shape
```

```python
df_files2['titulo'] = df_files2['titulo'].str.lower()
df_files2.head()
```

```python
df_files2['titulo'] = df_files2['titulo'].replace(" ", "")
df_files2.head()
```

```python
len(df_files2['titulo'])-len(df_files2['titulo'].drop_duplicates())
```

```python
df_files2.drop_duplicates(subset=['titulo'], inplace=True)
df_files2.shape
```

```python
df_files2 = df_files2[(df_files2['name_file'] != '1.pdf')]
df_files2.shape
```

```python
df_files2.head()
```

```python
df_files2[['a', 'b', 'c', 'd', 'e', 'f']] = df_files2['data_noticia_orig'].str.rsplit(' ', n=5, expand=True)
df_files2.head()
```

```python
df_files2 = df_files2[['name_file', 'data_noticia_orig', 'Publication_orig', 'titulo', 'Size_notice', 'Sumarization', 'a', 'b', 'c']]
df_files2.head()
```

```python
df_files2['b'] = df_files2['b'].replace(',', '', regex=True)
df_files2.head()
```

```python
def binary_map(feature):
    return feature.map({'May':'05', 'July':'07', 'September':'09', 'April':'04', 'February':'02', 'January':'01', 'June':'06', 'August':'08', 'November':'11', 'October':'10', 'March':'03', 'December':'12'})

## Encoding target feature
df_files2['d'] = df_files2[['a']].apply(binary_map)
df_files2.head()
```

```python
df_files2['e'] = df_files2['c'] + '-' + df_files2['d'] + '-' + df_files2['b']
df_files2['e'] = df_files2['e'].astype('str')
df_files2['data_noticia_corrig'] = pd.to_datetime(df_files2['e'])
df_files2['Ano'] = df_files2['data_noticia_corrig'].dt.year
df_files2.head()
```

```python
df_files2 = df_files2[['name_file', 'data_noticia_orig', 'data_noticia_corrig', 'Ano', 'Publication_orig', 'titulo', 'Size_notice', 'Sumarization']]
df_files2.head()
```
