```python
path_out_excel = '/drive/MyDrive/Doutorado_v2/Staging/'
```

```python
for a in range(1, 36):
  E01_file_reportagens['Size_notice_cum'] = E01_file_reportagens['Size_notice'].cumsum()
  E01_file_reportagens_STG = E01_file_reportagens[(E01_file_reportagens['Size_notice_cum'] <= 250000)]
  E01_file_reportagens = E01_file_reportagens[(E01_file_reportagens['Size_notice_cum'] > 250000)]

  b = str(a)
  name_file_df = 'file_reportagens_' + b + '.xlsx'
  file_out = os.path.join(path_out_excel, name_file_df)

  E01_file_reportagens_STG.to_excel(file_out, index=False)
```
