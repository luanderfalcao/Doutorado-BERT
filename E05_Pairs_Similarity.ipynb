```python
DF['N'] = 1
DF = DF[['name_file', 'N']]
```

```python
DF2 = pd.merge(DF, DF, on='N', suffixes=['','2']).query("name_file != name_file2")
DF2 = DF2[['name_file', 'name_file2', 'N']]
DF2 = DF2.rename(columns={'name_file': 'Col_X', 'name_file2': 'Col_Y'})
```

```python
path_out_excel = '/drive/MyDrive/Doutorado_v2/Staging/'
```

```python
%%time
for a in range(1, 209):
  DF2['N_cum'] = DF2['N'].cumsum()
  DF2_STG = DF2[(DF2['N_cum'] <= 50000)]
  DF2 = DF2[(DF2['N_cum'] > 50000)]

  b = str(a)
  name_file_df = 'Similary_Excel_' + b + '.xlsx'
  file_out = os.path.join(path_out_excel, name_file_df)

  DF2_STG.to_excel(file_out, index=False)
```
