```python
def sentiment_calculator(review):
    tokens = tokenizer.encode(review, return_tensors="pt", padding=True, truncation=True,max_length=100, add_special_tokens = True)
    result = model(tokens)
    return(int(torch.argmax(result.logits))+1)
```

```python
def chr_remove(old, to_remove):
    new_string = old
    for x in to_remove:
        new_string = new_string.replace(x, '')
    return new_string
```

```python
k = 1
name_file_df = summary_full_ajust.iloc[k]['name_file']
path_in = '/drive/MyDrive/Doutorado_v2/Files_in_PDF'
file_in = os.path.join(path_in, name_file_df)

text = ''
with fitz.open(file_in) as doc:
   for page in doc:
       text+= page.get_text()

page004 = text.split("\n")[4:]
page004 = repr(page004)
page004 = chr_remove(page004, "',")
page004 = chr_remove(page004, '"')
page004 = " ".join(page004.split())

summary_full_ajust.loc[k, 'text_full'] = page004
df = summary_full_ajust[['text_full']]
```

```python
%%time
for i in range(0, len(summary_full_ajust)):
  #i = 3

  name_file_df = summary_full_ajust.iloc[i]['name_file']
  path_in = '/drive/MyDrive/Doutorado_v2/Files_in_PDF'
  file_in = os.path.join(path_in, name_file_df)

  text = ''
  with fitz.open(file_in) as doc:
     for page in doc:
         text+= page.get_text()

  page004 = text.split("\n")[4:]
  page004 = repr(page004)
  page004 = chr_remove(page004, "',")
  page004 = chr_remove(page004, '"')
  page004 = chr_remove(page004, '[')
  page004 = chr_remove(page004, ']')
  page004 = " ".join(page004.split())

  summary_full_ajust.loc[i, 'text_full'] = page004

  text_sum = summary_full_ajust.iloc[i]['text_full']
  summary_full_ajust.loc[i, 'Sentiment'] = sentiment_calculator(text_sum)
  #del sentiment_calculator
  del text_sum
  #print(i)
```
