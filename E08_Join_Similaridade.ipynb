```python
file_A['Fx_Similarity'] = np.where(file_A['Similarity'] >= 0.95, 'Perfeita',
                          np.where(file_A['Similarity'] >= 0.75, 'Forte',
                          np.where(file_A['Similarity'] >= 0.5, 'Moderada',
                          np.where(file_A['Similarity'] >= 0.25, 'Fraca',
                          np.where(file_A['Similarity'] >= 0.1, 'Ínfima', 'Nula')))))
```

```python
file_B = pd.crosstab(file_A['Col_X'],
            file_A['Fx_Similarity'],
            values = file_A['Col_Y'],
            aggfunc='count', margins=False).replace(np.nan, 0).reset_index()
file_B.head(), file_B.shape
```

```python
file_B = file_B[['Col_X', 'Nula', 'Ínfima', 'Fraca', 'Moderada', 'Forte', 'Perfeita']]
file_B.head()
```

```python
file_B['Total'] = file_B['Ínfima'] + file_B['Nula'] + file_B['Fraca'] + file_B['Moderada'] + file_B['Forte'] + file_B['Perfeita']
file_B.head()
```

```python
file_B_teste = file_B[(file_B['Perfeita'] == 0)]
file_B_teste.shape
```

```python
variaveis2 = ['Nula', 'Ínfima', 'Fraca', 'Moderada', 'Forte', 'Perfeita']
```

```python
EDA_columns = []
EDA_compare = pd.DataFrame(columns = EDA_columns)
row_index = 0
for variaveis in (variaveis2):
    EDA_compare.loc[row_index,'Variável'] = variaveis
    EDA_compare.loc[row_index, 'Count'] = file_B[variaveis].count()
    EDA_compare.loc[row_index, 'Min'] = file_B[variaveis].min()
    EDA_compare.loc[row_index, 'Max'] = file_B[variaveis].max()
    EDA_compare.loc[row_index, 'Media'] = file_B[variaveis].mean()
    EDA_compare.loc[row_index, 'Mediana'] = file_B[variaveis].median()
    EDA_compare.loc[row_index, 'Count 0'] = (file_B[variaveis] == 0).sum()
    EDA_compare.loc[row_index, 'Count <> 0'] = (file_B[variaveis] != 0).sum()
    row_index+=1
EDA_compare
```
